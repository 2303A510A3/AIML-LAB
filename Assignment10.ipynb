Import the required Python packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.cluster import SpectralClustering
```

Create synthetic dataset of unlabeled blobs
Generate synthetic data with 3 centers (clusters) and 300 samples

```python
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

 Import KMeans class and fit the data

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

 Visualize the fitted data

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title("K-Means Clustering with Synthetic Data")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

part 2
Generate synthetic data

```python
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

Create and fit the KMeans model

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)
```

Plot the data points and cluster centers

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Cluster Centers')
plt.title("K-Means Clustering with Synthetic Data")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

Part 3

```python
X = np.array([[1, 2], [2, 3], [3, 2], [8, 8], [9, 10], [10, 9]])
```

```python
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels='kmeans', n_neighbors=3)
labels = model.fit_predict(X)
```

```python
# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.title('Spectral Clustering')
plt.show()
```
